```python
import sqlite3

def extract_schema_text(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = [row[0] for row in cursor.fetchall()]

    schema_chunks = []

    for table in tables:
        cursor.execute(f"PRAGMA table_info({table})")
        columns = cursor.fetchall()

        schema_text = f"Table: {table}\n"
        for col in columns:
            col_name = col[1]
            col_type = col[2]
            schema_text += f" - {col_name} ({col_type})\n"

        schema_chunks.append(schema_text.strip())

    conn.close()
    return schema_chunks
```

```python
schema_chunks = extract_schema_text("Chinook_Sqlite.sqlite")

print(f"Total tables found: {len(schema_chunks)}\n")

for i, chunk in enumerate(schema_chunks, 1):
    print(f" Table {i}:\n{chunk}")
    print("====\n")
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")
if schema_chunks:
    embeddings = model.encode(schema_chunks, show_progress_bar=True)
else:
    print("Warning: schema_chunks is empty. Embeddings will be empty as well.")
    embeddings = np.array([])


if embeddings.size > 0:
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(np.array(embeddings))
    faiss.write_index(index, "schema_index.faiss")

    with open("schema_chunks.txt", "w", encoding="utf-8") as f:
        for chunk in schema_chunks:
            f.write(chunk + "\n\n====\n\n")
else:
    print("Error: Embeddings are empty. Cannot proceed with indexing.")
```

```python
import gradio as gr
import sqlite3
import torch
import faiss
import numpy as np
import pandas as pd

from transformers import AutoTokenizer, AutoModelForCausalLM
from sentence_transformers import SentenceTransformer

# Paths
DB_PATH = "Chinook_Sqlite.sqlite"
SCHEMA_TXT = "schema_chunks.txt"
INDEX_PATH = "schema_index.faiss"

# Load embedding model and schema
embed_model = SentenceTransformer("all-MiniLM-L6-v2")
index = faiss.read_index(INDEX_PATH)
with open(SCHEMA_TXT, "r", encoding="utf-8") as f:
    schema_chunks = f.read().split("\n\n====\n\n")

# Load Mistral model
model_name = "mistralai/Mistral-7B-Instruct-v0.2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
)

# --- Core Functions ---

def retrieve_chunks(query, k=3):
    embedding = embed_model.encode([query])
    _, indices = index.search(np.array(embedding), k)
    return [schema_chunks[i] for i in indices[0]]

def build_prompt(context, question, mode):
    if mode == "sql":
        return f"""<s>[INST]
You are an expert SQL assistant. Based on the following database schema:

{context}

Write an SQL query to answer this question:
\"\"\"{question}\"\"\"

SQL:
[/INST]"""
    else:
        return f"""<s>[INST]
You are a helpful assistant. Based on the following database schema:

{context}

Answer the following question in natural language (not SQL):
\"\"\"{question}\"\"\"

Answer:
[/INST]"""

def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            temperature=0.0,
            pad_token_id=tokenizer.eos_token_id
        )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

def ask_question(question, mode, history):
    if not question.strip():
        return "❌ Please enter a valid question.", None, history, ""

    # RAG process
    chunks = retrieve_chunks(question)
    context = "\n\n".join(chunks)
    prompt = build_prompt(context, question, mode)
    response = generate_response(prompt)

    # Save to chat memory
    history.append({"question": question, "response": response})

    if mode == "sql":
        sql_start = response.lower().find("sql:")
        sql_code = response[sql_start + 4:].strip() if sql_start != -1 else response.strip()

        try:
            conn = sqlite3.connect(DB_PATH)
            df = pd.read_sql_query(sql_code, conn)
            return sql_code, df.head(10), history, context
        except Exception as e:
            return f"❌ Error executing SQL:\n{e}\n\n🔹 Model Response:\n{response}", None, history, context
    else:
        return response, None, history, context

def display_history(history):
    return "\n\n".join([f"🟩 Q: {h['question']}\n🟦 A: {h['response']}" for h in history])

# --- Gradio UI ---
with gr.Blocks(title="Chinook RAG QA") as demo:
    gr.Markdown("## 💡 Ask Questions About the Chinook Database")

    with gr.Row():
        question = gr.Textbox(label="Enter your question", placeholder="e.g. List all albums released by Canadian artists.")
        mode = gr.Radio(["sql", "natural"], value="sql", label="Response Mode", info="Choose how the model should answer")

    submit_btn = gr.Button("Generate Response")

    with gr.Row():
        output_text = gr.Textbox(label="Model Response", lines=4)
        output_df = gr.Dataframe(label="SQL Query Result (if applicable)", visible=True)

    with gr.Row():
        explain_box = gr.Textbox(label="Retrieved Schema Context", lines=10, interactive=False)

    with gr.Row():
        history_box = gr.Textbox(label="Chat History", lines=10, interactive=False)

    chat_history = gr.State([])

    # Main response generation
    submit_btn.click(
        fn=ask_question,
        inputs=[question, mode, chat_history],
        outputs=[output_text, output_df, chat_history, explain_box]
    )

    # Chat history display
    submit_btn.click(display_history, inputs=[chat_history], outputs=[history_box])

demo.launch()
```
